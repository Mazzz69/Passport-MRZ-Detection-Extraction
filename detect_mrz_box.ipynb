```python
import os
import cv2
import matplotlib.pyplot as plt

SAMPLES_DIR = os.path.join(os.getcwd(), "samples")

def load_image(image_name):
    image_path = os.path.join(SAMPLES_DIR, image_name)

    if not os.path.exists(image_path):
        raise ValueError("image not found in samples")

    img_rgb = cv2.imread(image_path, cv2.IMREAD_COLOR_RGB)

    if img_rgb is None: 
        raise ValueError("image is null")

    return img_rgb

# passport1.jpg
# function to test load_image()
if __name__ == "__main__":
    image_name = input("Insert image name: ").strip()
    image = load_image(image_name)

    cv2.imwrite(image_name, image)
    print(image.shape)
    plt.imshow(image)
    plt.axis("off")
```
